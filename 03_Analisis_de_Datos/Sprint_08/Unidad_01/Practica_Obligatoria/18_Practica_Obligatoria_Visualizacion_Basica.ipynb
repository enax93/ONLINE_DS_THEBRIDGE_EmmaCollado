### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada gráfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrón, con trasparencia)

```python
ciudades_pobladas = df_cities.nlargest(5, "population_total")

ciudades = ciudades_pobladas["city"]
poblaciones = ciudades_pobladas["population_total"] / 1000

plt.bar(ciudades, poblaciones, color=["blue","gray","red","olive","brown"], alpha=0.5)


for ciudad, poblacion in zip(ciudades, poblaciones):
    plt.text(ciudad, poblacion, f'{poblacion * 1000:,.0f}', ha='center')

plt.title("Ciudades más pobladas de California")
plt.xlabel("Ciudades")
plt.ylabel("Población (miles)")
plt.ticklabel_format(style="plain",axis="y")
plt.yticks(range(0, 6000,1000));
```

2. Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamaño 18, color azul acero)

```python
plt.hist(df_cities['area_total_km2'], bins=50, color='steelblue', label="Hist Area")

font = {
    'fontname': 'Bradley Hand ITC', 
    'size': 18
}

plt.xlabel("Superficie ($km^2$)", fontdict=font)
plt.ylabel("Número de ciudades", fontdict=font)
plt.xticks(range(0, 1400,200))
plt.legend();
```

3. El scatter de área de agua de un ciudad con su poblacion.

```python
area_agua = df_cities["area_water_km2"]
poblacion = df_cities["population_total"] / 1000

max_pob = poblacion.idxmax()
x_max_pob = area_agua[max_pob]
y_max_pob = poblacion[max_pob]

max_agua_idx = area_agua.idxmax()
x_max_agua = area_agua[max_agua_idx]
y_max_agua = poblacion[max_agua_idx]

plt.scatter(area_agua, poblacion, color='green', alpha=0.5)

plt.annotate("Outlier", xy=(x_max_pob, y_max_pob),
             xytext=(60, 3500),
             arrowprops={"arrowstyle": "->"})

plt.annotate("Outlier", xy=(x_max_agua, y_max_agua),
             xytext=(60, 3500),
             arrowprops={"arrowstyle": "->"})

plt.title("Población por Kilómetro Cuadrado de Agua")
plt.xlabel("Área de Agua en la Ciudad ($Km^2$)")
plt.ylabel("Población (Miles de Habitantes)")
plt.xticks(range(0, 140, 20))
plt.yticks(range(0, 5000, 1000));
```

4. El boxplot de la altura de las ciudades.

```python
altura_ciudades = df_cities['elevation_m']

plt.boxplot(altura_ciudades, vert=False, patch_artist=True, boxprops=dict(facecolor='red'))

plt.title("Diagrama de caja de la Altitud de las ciudades de California")
plt.xlabel("Altitud (m)");
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

ciudades_pobladas = df_cities.nlargest(5, "population_total")
ciudades = ciudades_pobladas["city"]
poblaciones = ciudades_pobladas["population_total"] / 1000

axs[0, 0].bar(ciudades, poblaciones, color=["blue", "gray", "red", "olive", "brown"], alpha=0.5)
for ciudad, poblacion in zip(ciudades, poblaciones):
    axs[0, 0].text(ciudad, poblacion, f'{poblacion * 1000:,.0f}', ha='center')
axs[0, 0].set_title("Ciudades más pobladas de California")
axs[0, 0].set_xlabel("Ciudades")
axs[0, 0].set_ylabel("Población (miles)")
axs[0, 0].ticklabel_format(style="plain", axis="y")
axs[0, 0].set_yticks(range(0, 6000, 1000))

axs[0, 1].hist(df_cities['area_total_km2'], bins=50, color='steelblue', label="Hist Area")
axs[0, 1].set_xlabel("Superficie ($km^2$)")
axs[0, 1].set_ylabel("Número de ciudades")
axs[0, 1].set_xticks(range(0, 1400, 200))
axs[0, 1].legend()

area_agua = df_cities["area_water_km2"]
poblacion = df_cities["population_total"] / 1000
axs[1, 0].scatter(area_agua, poblacion, color='green', alpha=0.5)
max_pob = poblacion.idxmax()
x_max_pob = area_agua[max_pob]
y_max_pob = poblacion[max_pob]
max_agua_idx = area_agua.idxmax()
x_max_agua = area_agua[max_agua_idx]
y_max_agua = poblacion[max_agua_idx]
axs[1, 0].annotate("Outlier", xy=(x_max_pob, y_max_pob), xytext=(60, 3500), arrowprops={"arrowstyle": "->"})
axs[1, 0].annotate("Outlier", xy=(x_max_agua, y_max_agua), xytext=(60, 3500), arrowprops={"arrowstyle": "->"})
axs[1, 0].set_title("Población por Kilómetro Cuadrado de Agua")
axs[1, 0].set_xlabel("Área de Agua en la Ciudad ($Km^2$)")
axs[1, 0].set_ylabel("Población (Miles de Habitantes)")
axs[1, 0].set_xticks(range(0, 140, 20))
axs[1, 0].set_yticks(range(0, 5000, 1000))

altura_ciudades = df_cities['elevation_m']
axs[1, 1].boxplot(altura_ciudades, vert=False, patch_artist=True, boxprops=dict(facecolor='red'))
axs[1, 1].set_title("Diagrama de caja de la Altitud de las ciudades de California")
axs[1, 1].set_xlabel("Altitud (m)")

plt.tight_layout();

```
