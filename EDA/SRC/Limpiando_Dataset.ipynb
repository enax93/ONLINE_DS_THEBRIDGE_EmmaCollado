```python
# definimos el capture_rate para el pokemon anterior, Minior es un pokemon que vienen en forma meteorito por defecto, al recibir daño por encima del 50% cambia a la forma nucleo.
# Segun su informacion el capture_rate del nucleo es de 255, por tal razon, utilizaremos este valor para dicho pokemon.
df_1.loc[773,"capture_rate"] = 255
```

```python
# Revisamos el capture_rate nuevamente para comprobar el cambio
df_1.capture_rate.value_counts()
```

```python
# cambiamos el tipo de dato de la columna completa a numeros enteros
df_1["capture_rate"] = df_1["capture_rate"].apply(lambda x: int(x))
```

```python
# Cambiamos el tipo de dato de una lista de string a una lista de elementos 
df_1["abilities"] = df_1.apply(lambda x: literal_eval(x["abilities"]), axis=1)
```

```python
# Creamos una columna con el numero de habilidades que puede tener cada pokemon
df_1["num_abilities"] = df_1.apply(lambda x: len(x['abilities']), axis=1)
```

```python
# creamos una columna tipo en donde reflejamos la combinacion de tipo1 y tipo2
df_1['type'] = df_1.apply(lambda x: x['type1'] if pd.isnull(x['type2']) else f'{x["type1"]}_{x["type2"]}', axis=1)
```

```python
valores_nulos = df_1[df_1["weight_kg"].isnull()][["name","weight_kg","height_m"]]
valores_nulos
```

```python
# Filtrar las filas con valores nulos en "weight_kg" y "height_kg"
filas_nulas = df_1[df_1["weight_kg"].isnull() & df_1["height_m"].isnull()]

# Seleccionar solo las columnas "pokedex_number", "weight_kg" y "height_m"
datos_faltantes = filas_nulas[["pokedex_number","name", "weight_kg", "height_m"]]

# Guardar los datos faltantes en un nuevo archivo CSV
datos_faltantes.to_csv("./data/pkmn_con_nulos.csv", index=False)
```

```python
# Cargar el archivo CSV con la información completa
pkmn_con_nulos_completados = pd.read_csv("./data/pkmn_con_nulos_completados.csv") 

# Combinar los conjuntos de datos utilizando el número de Pokedex como clave
df_1_actualizado = pd.merge(df_1, pkmn_con_nulos_completados, on='pokedex_number', how='left', suffixes=('', '_nuevo'))

# Actualizar los valores faltantes en el dataset original con los valores del dataset actualizado
df_1_actualizado['weight_kg'] = df_1_actualizado['weight_kg_nuevo'].fillna(df_1_actualizado['weight_kg'])
df_1_actualizado['height_m'] = df_1_actualizado['height_m_nuevo'].fillna(df_1_actualizado['height_m'])

# Eliminar las columnas adicionales
df_1_actualizado.drop(["name_nuevo", 'weight_kg_nuevo', 'height_m_nuevo'], axis=1, inplace=True)
```

```python
porcentaje_nulo = df_1_actualizado.isnull().mean() * 100
print(porcentaje_nulo[porcentaje_nulo > 0])
```

```python
# Guardar el dataset actualizado en un nuevo archivo CSV
df_1_actualizado.to_csv('./data/pokemon_actualizado.csv', index=False)
```
